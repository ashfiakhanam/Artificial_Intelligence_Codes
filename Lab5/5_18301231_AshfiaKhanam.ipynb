```python
#importing necessary libraries
import pandas as pd
import numpy as np
```

```python
#Reading data
wine = pd.read_csv("wine.csv")
wine.head(20)
```

```python
#Printing number of row and columns
wine.shape
```

```python
#checking nulls
wine.isnull()
```

```python
#checking number of nulls in a column 
wine.isnull().sum()
```

```python
#Encoding categorical variables:
wine.info()
```

```python
#printing catagorical feature
wine['quality'].unique()
```

```python
#Label Encoder
from sklearn.preprocessing import LabelEncoder
lenc = LabelEncoder()

# Apply the encoding
wine["quality_enc"] = lenc.fit_transform(wine["quality"])

# Compare the two columns
print(wine[["quality", "quality_enc"]].head())
```

```python
#Split into features and labels
wine_df = wine.drop(['quality'], axis = 1)
wine_df.head()
wine_corr = wine_df.corr()
wine_corr
```

```python
import seaborn as sns
sns.heatmap(wine_corr, cmap = 'YlGnBu')
```

```python
# Take a minute to find the column where the correlation value is greater than 0.75 at least twice
to_drop = [ "residual sugar", "free sulfur dioxide", "chlorides", "pH"]

# Drop that column from the DataFrame
wine_df = wine_df.drop(to_drop, axis=1)
wine_df.head(5)
```

```python
#preparing training set

x = wine_df.drop("quality_enc", axis = 1)
y =  wine_df[["quality_enc"]]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.25, random_state = 45)
print(X_train.shape)
print(X_test.shape)
```

```python
#0-1 scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
#Transforming Data
X_train_scaled = scaler.transform(X_train)
```

```python
print("per-feature minimum before scaling:\n {}".format(X_train.min(axis=0)))
print("per-feature maximum before scaling:\n {}".format(X_train.max(axis=0)))
```

```python
print("per-feature minimum after scaling:\n {}".format(
    X_train_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(
    X_train_scaled.max(axis=0)))
```

```python
# transform test data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#performing logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression()

#train the model
lr.fit(X_train_scaled, Y_train) 
predictions = lr.predict(X_test_scaled)

print(predictions)



```

```python
print("Accuracy:",accuracy_score(Y_test, predictions))
```
